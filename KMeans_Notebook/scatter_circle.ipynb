```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import matplotlib.transforms as transforms
import pandas as pd
```

```python
file_to_read="Banknote.csv"
df_banknote=pd.read_csv(file_to_read)
```

```python
min_data=np.min(df_banknote,0)
max_data=np.max(df_banknote,0)
normed_data=(df_banknote- min_data)/ (max_data- min_data)
normed_data.head()
```

```python

sdata=  np.array(normed_data)

fig, ax = plt.subplots()

colours =['magenta', 'blue']

V1_mean = np.mean(sdata[:,0])
print ("media en V1: ", V1_mean)
V2_mean = np.mean(sdata[:,1])
print ("media en V2: \n ", V2_mean)
cov = np.cov(sdata[:,0], sdata[:,1])
print ("covariance : \n", cov)
nstd = (np.std(sdata))* 2
ax.scatter(sdata[:,0], sdata[:,1],  color="magenta",
               label="circle with radio= square standard deviation", s=5)
e = Circle (xy=(V1_mean, V2_mean), radius=nstd, alpha =0.25)

ax.add_artist(e)

ax.set_xlabel('V1')
ax.set_ylabel('V2')
ax.legend(loc='upper left', scatterpoints=1)
plt.show()
```
